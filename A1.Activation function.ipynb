```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Generate input values
x = np.linspace(-10, 10, 400)

# Calculate outputs for each activation function
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)

# Plot each activation function on a separate subplot
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y_sigmoid, color='blue')
axs[0, 0].set_title('Sigmoid Activation Function')
axs[0, 0].set_xlabel('Input')
axs[0, 0].set_ylabel('Output')
axs[0, 0].grid(True)

axs[0, 1].plot(x, y_tanh, color='green')
axs[0, 1].set_title('Tanh Activation Function')
axs[0, 1].set_xlabel('Input')
axs[0, 1].set_ylabel('Output')
axs[0, 1].grid(True)

axs[1, 0].plot(x, y_relu, color='red')
axs[1, 0].set_title('ReLU Activation Function')
axs[1, 0].set_xlabel('Input')
axs[1, 0].set_ylabel('Output')
axs[1, 0].grid(True)

axs[1, 1].plot(x, y_leaky_relu, color='purple')
axs[1, 1].set_title('Leaky ReLU Activation Function')
axs[1, 1].set_xlabel('Input')
axs[1, 1].set_ylabel('Output')
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
